```python
from textblob import TextBlob
import pandas as pd
```

```python
# Open a file: file
f = open("D://ISB_CBA_COURSE/CAMPUS/TERM1_SECOND_Residency/Assignments/TextAnalytics/Submission/Nokia_Lumia_Reviews_en.txt", encoding = "utf-8")

# Get the file Content
en_text = f.readlines()

en_blob = TextBlob(''.join(en_text))

print(en_blob)

# close the file
f.close()
```

```python
hi_blob = en_blob.translate(from_lang="en", to='hi')

print(hi_blob)
```

```python
es_blob = en_blob.translate(from_lang="en", to='es')

print(es_blob)
```
